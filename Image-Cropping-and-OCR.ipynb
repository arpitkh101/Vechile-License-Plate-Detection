```python
import pandas as pd
import urllib
import matplotlib.pyplot as plt
import numpy as np
import cv2
import glob
import os
import time
from PIL import Image
import pytesseract
```

```python
df = pd.read_csv(r"C:\Users\ArpitKh4101\NumberPlates.csv")
```

```python
df.head()
```

```python
WIDTH = 224
HEIGHT = 224
CHANNEL = 3
```

```python
def show_img(index):
    image = cv2.imread(r"C:\Users\ArpitKh4101\Downloads\INDIAN/" + df["image_name"].iloc[index])
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, dsize=(WIDTH, HEIGHT))

    tx = int((df["centre_x"].iloc[index] - df["width"].iloc[index]/2)* WIDTH )
    ty = int((df["centre_y"].iloc[index] - df["height"].iloc[index]/2) * HEIGHT)
    bx = int((df["centre_x"].iloc[index] + df["width"].iloc[index]/2) * WIDTH)
    by = int((df["centre_y"].iloc[index] + df["height"].iloc[index]/2) * HEIGHT)
    
   
    
    image = cv2.rectangle(image, (tx, ty), (bx, by), (0, 0, 255), 1)
    plt.imshow(image)

    plt.show()
    
```

```python
def crop_img(index):
    img = cv2.imread(r"C:\Users\ArpitKh4101\Downloads\INDIAN/" + df["image_name"].iloc[index])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, dsize=(WIDTH, HEIGHT))
    
    tx = int((df["centre_x"].iloc[index] - df["width"].iloc[index]/2)* WIDTH )
    ty = int((df["centre_y"].iloc[index] - df["height"].iloc[index]/2) * HEIGHT)
    bx = int((df["centre_x"].iloc[index] + df["width"].iloc[index]/2) * WIDTH)
    by = int((df["centre_y"].iloc[index] + df["height"].iloc[index]/2) * HEIGHT)
    
    
    crop_img = img[ty:by, tx:bx]
    
    print(pytesseract.image_to_string(crop_img))
    
    
    plt.imshow(crop_img)
    plt.show()
```

```python
show_img(6)
crop_img(6)
```
